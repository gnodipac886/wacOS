```python
from scipy import misc
import numpy as np
import imageio

np.set_printoptions(formatter={'int':hex})
```

```python
s = input("Please type file name here(with .png): ")
name = s.split('.png')[0]
arr = np.array(imageio.imread(s))
```

```python
rgb_565 = np.array([], dtype=np.uint16)
# Red - 5, Green - 6, Blue - 5
# RRRRRGGGGGGBBBBB
for i in range(arr.shape[0]):
    for j in range(arr.shape[1]):
#         red = (((arr[i, j, 0] & 0xF8) >> 3) + int((arr[i, j, 0] & 0x07) > 3)) << 11
#         green = (((arr[i, j, 1] & 0xFC) >> 2) + int((arr[i, j, 1] & 0x03) > 1)) << 5
#         blue = (((arr[i, j, 2] & 0xF8)) >> 3) + int((arr[i, j, 2] & 0x07) > 0)
        try:
            red = (arr[i, j, 0] >> 3) << 11
            green = (arr[i, j, 1] >> 2) << 5
            blue = (arr[i, j, 2] >>3)
            rgb_565 = np.append(rgb_565, np.uint16(red | green | blue))
        except:
            color = arr[i, j] if arr[i, j] == 0 else 0xff
            rgb_565 = np.append(rgb_565, np.uint16(arr[i, j]))
#         rgb_565.append(red | green | blue)
# rgb_565 = np.array(rgb_565, dtype=np.uint8)
```
